```python
import pandas as pd
df = pd.read_csv('aging_model/age_des_vins_v6.csv')
end_df = pd.read_csv('aging_model/age_des_vins_end.csv')
areas = pd.read_csv('aging_model/areas.csv')
```

```python
df['area']=df['area'].map(lambda x: x.upper())
df['cru']=df['cru'].map(lambda x: x.upper())
```

```python
df
```

```python
areas['name'] = areas['name'].map(lambda x: x.upper())
areas['true_area_id']=areas['id']
```

```python
first_join = areas.merge(df, how='outer', right_on='cru', left_on='name')
reste = first_join[first_join.area.isna()][['id','name']]
reste
```

```python
res_1 = first_join[(first_join['name'].isna()==False)&(first_join['area'].isna()==False)]
```

```python
sec_join = reste.merge(df, how='outer', left_on='name', right_on='area')
reste_2 = sec_join[sec_join.area.isna()][['id','name']]
reste_2
```

```python
res_2 = sec_join[(sec_join['name'].isna()==False)&(sec_join['area'].isna()==False)]
```

```python
end_df = pd.read_csv('aging_model/age_des_vins_end.csv', delimiter=";")
end_df.name=end_df.name.map(lambda x: x.upper())
end_df['area'] = end_df['name']
end_df.drop( inplace=True, labels='name', axis=1)
```

```python
end_df[end_df['id']==71]
```

```python
third_join = reste_2.merge(end_df, how='outer', left_on='name', right_on='area')
third_join
```

```python
res_3 = third_join[(third_join['name'].isna()==False)&(third_join['area'].isna()==False)]
res_3 = res_3[res_3.color.isna()==False]
```

```python
res_3
```

```python
final = pd.concat([res_1,res_2,res_3]).merge(areas, how='inner', on='name')
final
```

```python
result = final[['age_min','age_max','true_area_id','t_min','t_max','color','comments','glass_type','type']]
result.to_csv('age_des_vins_v8.csv', index=False)
```
